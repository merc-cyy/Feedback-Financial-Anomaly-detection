```python


# Define a function to detect anomalies using Z-score analysis
def detect_anomalies(data_series, threshold=3):
    z_scores = (data_series - data_series.mean()) / data_series.std()
    return data_series.index[np.abs(z_scores) > threshold]

# Ensure the 'Date' column is parsed as datetime for proper time series analysis
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with missing or non-convertible dates
df = df.dropna(subset=['Date']).sort_values(by='Date')

# Select the columns for anomaly detection
selected_columns = ['VIX Index', 'Cl1 Comdty', 'XAU BGNL Curncy', 'DXY Curncy', 'BDIY Index', 'USGG30YR Index','ECSURPUS Index']
df_subset = df[['Date'] + selected_columns].set_index('Date')

# Identify anomalies for each selected column
anomalies = {col: detect_anomalies(df_subset[col]) for col in selected_columns}

# Plot each metric with anomalies highlighted
fig, axes = plt.subplots(len(selected_columns), 1, figsize=(14, 14), sharex=True)
for i, col in enumerate(selected_columns):
    ax = axes[i]
    ax.plot(df_subset.index, df_subset[col], label=f"{col} Trend", color='blue', alpha=0.7)
    ax.scatter(
        anomalies[col],
        df_subset.loc[anomalies[col], col],
        color='red',
        label="Anomalies",
        zorder=5
    )
    ax.set_title(f"Anomalies in {col}")
    ax.legend(loc='upper left')
    ax.grid(True)

plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import zscore
```

```python
z_scores_df = df.select_dtypes(include=['float64', 'int64']).apply(zscore)
df_with_zscores = df.copy()
for col in z_scores_df.columns:
    df_with_zscores[f"{col}_zscore"] = z_scores_df[col]
```

```python
df_with_zscores.head()
```

```python
anomaly_threshold = 3
anomalies = (z_scores_df.abs() > anomaly_threshold)
anomalies_summary = {}
for col in z_scores_df:
    anomalies_summary[col] = df[anomalies[col]]

anomalies_combined = df[anomalies.any(axis=1)]

# View results
print(f"Anomalies detected in each column:")
for col, anomaly in anomalies_summary.items():
    print(f"{col}: {len(anomaly)} anomalies detected.")

print("\nCombined anomalies (sample):")
anomalies_combined.head()
```

```python
# Define thresholds for specific columns (example values)
thresholds = {
    'VIX Index': {'min': 0, 'max': 30},  # VIX > 30 indicates high market stress
    'Cl1 Comdty': {'min': 20, 'max': 150},  # Crude oil prices range
    'XAU BGNL Curncy': {'min': 1000, 'max': 2000},  # Gold price range
    'DXY Curncy': {'min': 80, 'max': 120},  # Dollar index range
    'BDIY Index': {'min': 500, 'max': 5000}  # Shipping index range
}

# Initialize an empty DataFrame to store anomalies
anomalies_combined = pd.DataFrame()

# Apply thresholds and find anomalies
for col, bounds in thresholds.items():
    anomalies = df[(df[col] < bounds['min']) | (df[col] > bounds['max'])]
    anomalies_combined = pd.concat([anomalies_combined, anomalies])

# Remove duplicate rows (in case multiple columns flagged the same row)
anomalies_combined = anomalies_combined.drop_duplicates()

# View anomalies
print(f"Detected anomalies:\n{anomalies_combined}")

# Save anomalies to CSV (optional)
anomalies_combined.to_csv('anomalies_detected.csv', index=False)
```

```python


# Function to visualize anomalies
def plot_anomalies(df, thresholds, selected_columns):
    fig, axes = plt.subplots(len(selected_columns), 1, figsize=(14, 10), sharex=True)
    for i, col in enumerate(selected_columns):
        ax = axes[i]

        # Plot the full data series
        ax.plot(df['Date'], df[col], label=f"{col} Trend", color='blue', alpha=0.7)

        # Highlight anomalies
        anomalies = df[(df[col] < thresholds[col]['min']) | (df[col] > thresholds[col]['max'])]
        ax.scatter(anomalies['Date'], anomalies[col], color='red', label="Anomalies", zorder=5)

        # Add threshold lines
        ax.axhline(y=thresholds[col]['min'], color='green', linestyle='--', label='Min Threshold')
        ax.axhline(y=thresholds[col]['max'], color='orange', linestyle='--', label='Max Threshold')

        ax.set_title(f"Anomalies in {col}")
        ax.legend(loc='upper left')
        ax.grid(True)

    plt.xlabel("Date")
    plt.tight_layout()
    plt.show()

# Define columns to visualize
selected_columns = ['VIX Index', 'Cl1 Comdty', 'XAU BGNL Curncy', 'DXY Curncy', 'BDIY Index']

# Plot anomalies
plot_anomalies(df, thresholds, selected_columns)
```

## Machine Learning model approach (Isolation Forest)

```python
from sklearn.ensemble import IsolationForest
import joblib
```

```python
#Prepare data for isolation
selected_columns = ['VIX Index', 'Cl1 Comdty', 'XAU BGNL Curncy', 'DXY Curncy', 'BDIY Index']
df_subset = df[selected_columns].dropna()
```

```python
#Train Isolation forest
isolation_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
df_subset['anomaly_score'] = isolation_forest.fit_predict(df_subset)
```

```python
#Save the trained Model
joblib.dump(isolation_forest, 'isolation_forest_model.pkl')
print("Model saved successfully.")
```

```python
df['Anomaly'] = -1
df.loc[df_subset.index, 'Anomaly'] = df_subset['anomaly_score']

#Visualize Anomalies
def plot_isolation_forest_anomalies(df, selected_columns):
  fig, axes = plt.subplots(len(selected_columns), 1, figsize=(14, 10), sharex=True)
  for i, col in enumerate(selected_columns):
        ax = axes[i]

        # Plot full data series
        ax.plot(df['Date'], df[col], label=f"{col} Trend", color='blue', alpha=0.7)

        # Highlight anomalies
        anomalies = df[df['Anomaly'] == -1]
        ax.scatter(anomalies['Date'], anomalies[col], color='red', label='Anomalies', zorder=5)

        ax.set_title(f"Isolation Forest Anomalies in {col}")
        ax.legend(loc='upper left')
        ax.grid(True)

  plt.xlabel("Date")
  plt.tight_layout()
  plt.show()
```

```python
# Plot anomalies
plot_isolation_forest_anomalies(df, selected_columns)
```

```python
#Load the trained model
isolation_forest = joblib.load('isolation_forest_model.pkl')
print("Model loaded successfully.")

```

```python
# Prepare new data for anomaly detection
new_data = pd.DataFrame({
    'VIX Index': [25, 40, 15],
    'Cl1 Comdty': [120, 180, 30],
    'XAU BGNL Curncy': [1800, 900, 1100],
    'DXY Curncy': [95, 130, 80],
    'BDIY Index': [3000, 100, 6000]
})

# Predict anomalies on new data
new_data['Anomaly_Score'] = isolation_forest.predict(new_data)

# Identify anomalies
new_anomalies = new_data[new_data['Anomaly_Score'] == -1]
print("Anomalies detected:")
print(new_anomalies)
```
